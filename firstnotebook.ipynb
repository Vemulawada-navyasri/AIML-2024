```python
people=["navyasri","srija","kavya","tina","swapna","raghu","raju"]
age=[1,2,3,4,5,6,7]
weight=[34,56,43,44,55,43,78]
height=[34,56,46,32,67,23,66]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(weight, height)
plt.title("relation between weight and height")
plt.ylabel("height in meter")
plt.xlabel("weight in kgs")
plt.show()
```

```python
plt.scatter(age, height)
plt.title("relation between weight and height")
plt.ylabel("height in meter")
plt.xlabel("age in integer")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.title("people weight in kgs",fontsize=16,fontstyle="italic")
plt.bar(x=people,height=weight,width=0,color='orange',edgecolor='red',alpha=0.6)
plt.xlabel("people",fontsize=12)
plt.ylabel("weight",fontsize=15)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,5))
plt.hist(weight,color='red',edgecolor='blue',alpha=0.75,bins=5)
plt.xlabel("histogram",fontsize=18)
plt.ylabel("weight",fontsize=15)
plt.show()
```
